```python
Income_2011= salaries_df.loc[salaries_df['Year']==2011]['TotalPayBenefits'].median()
Income_2012= salaries_df.loc[salaries_df['Year']==2012]['TotalPayBenefits'].median()
Income_2013= salaries_df.loc[salaries_df['Year']==2013]['TotalPayBenefits'].median()
Income_2014= salaries_df.loc[salaries_df['Year']==2014]['TotalPayBenefits'].median()
sns.barplot(x=['2011', '2012', '2013', '2014'], y=[Income_2011, Income_2012, Income_2013, Income_2014])
```

```python
salaries_df['JobTitle'].unique().shape
```

```python
salaries_df['JobTitle'].value_counts().nlargest(100)
```

```python
a= salaries_df['JobTitle'].value_counts().head(10).index.tolist()
avg= []
for i in a:
    b= salaries_df.loc[salaries_df['JobTitle']==i]['TotalPay'].median()
    avg.append(b)

print(avg)    

sns.barplot(x=a, y=avg, palette='viridis', alpha=0.6)
plt.xticks(rotation=90)
plt.xlabel('Job Titles', fontsize=16)
plt.ylabel("Avg TotalPay", fontsize=16)
plt.title("Avg TotalPay for top 10 available jobs")
plt.show()
```

```python
Years= [2011, 2012, 2013, 2014]
for i in Years:
    a= salaries_df.loc[(salaries_df['Year']==i)].sort_values(by='TotalPay', ascending=False).head(1)
    print('In {} the highest earned Employee is {} with total income {}'.format(i, a['EmployeeName'].values[0], 
                                                                                       a['TotalPay'].values[0]))
```
